# Project: Data visualazation - California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.available
plt.style.use('seaborn-whitegrid')
```

```python
cities = pd.read_csv('C:/Users/Admin/Desktop/Getting Knowing Data/california_cities.csv')
cities.head()

cities['area_total_km2'].sort_values(ascending = False)
```

```python
# Extract Latd = V  and Longd = Kinh 
lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']

plt.style.use('seaborn')
plt.figure(figsize=(8,6))
# Plot using Pyplot API
plt.scatter (lon, lat,
             c=np.log10(population), cmap='viridis',
             s=area, linewidths=0, alpha=0.5)
plt.axis('equal')
plt.xlabel('longtitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3,7);

#Create a legend for cities's sizes
area_range = [50,100,300,500]
for area in area_range:
    plt.scatter([],[], s = area, c = 'k', alpha=0.4,
                label=str(area) + 'km$^2$')
plt.legend(labelspacing=1,title='City A')

plt.title('Calfornia Cities: Population and Area Distribution');
```
